```python
import os
from google.colab import userdata
os.environ["OPENAI_API_KEY"] = userdata.get('openai_API_key')
```

```python
from openai import AsyncOpenAI
import openai
import asyncio
from tqdm.notebook import tqdm

client_async = AsyncOpenAI()

async def generate_response_async(prompt):
    messages = [{"role": "user", "content": prompt}]

    response = await client_async.chat.completions.create(
        model="gpt-3.5-turbo",
        messages=messages,
        timeout=1000
    )
    return response.choices[0].message.content


examples = ['How to make sushi?', 'How to make pizza?'] * 100

async def main():
    # Prepare prompts for async API calls
    prompts = [example for example in examples]

    # Create tasks for each prompt
    tasks = [generate_response_async(p) for p in prompts]

    # Use tqdm to show progress as each task completes
    results = []
    for task in tqdm(asyncio.as_completed(tasks), total=len(tasks)):
        results.append(await task)

    return results

# Run the main asynchronous function
# Use asyncio.run to run the asynchronous main function in a blocking way
results = await main()

# Print results
print(results)
```
